```python
from mem0 import Memory
from dotenv import load_dotenv

load_dotenv()



config= {
    "vector_store": {
        "provider": "chroma",
        "config": {
            "collection_name": "test",
            "path": "db",
            
        }
    },
    "llm": {
        "provider": "gemini",
        "config": {
            "model": "gemini-2.0-flash-001",
            "temperature": 0.2,
            "max_tokens": 2000,
            "top_p": 1.0
        }
    },
    "embedder": {
        "provider": "gemini",
        "config": {
            "model": "models/text-embedding-004",
        }
    }
}

m = Memory.from_config(config)
```

```python
messages = [
    {"role": "user", "content": "I'm planning to watch a movie tonight. Any recommendations?"},
    {"role": "assistant", "content": "How about thriller movies? They can be quite engaging."},
    {"role": "user", "content": "I'm not a big fan of thriller movies but I love sci-fi movies."},
    {"role": "assistant", "content": "Got it! I'll avoid thriller recommendations and suggest sci-fi movies in the future."}
]
m.add(messages, user_id="john")
```

```python
m.add([
    {"role": "user", "content": "Hi, I'm Vinod. I run FutureSmart AI, where we build custom AI solutions."}
], user_id="vinod")

m.add([
    {"role": "user", "content": "I love building RAG and AI Agent solutions that actually work in production."}
], user_id="vinod", metadata={"category": "preferences"})
```

```python
related = m.search("what do I build", user_id="vinod")
related
```

```python
all_memories = m.get_all(user_id="vinod")
memories = [mem['memory'] for mem in all_memories['results']]
print(memories)
```
